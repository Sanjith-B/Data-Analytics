Region Based Grouping Execution

```python
import os
import pandas as pd

# Define the folder containing datasets and the output folder
dataset_folder = "/content/drive/MyDrive/Demand Forecasting (Prince Pharma)/Engineered Data/Forecast Files/Shop Level"  # Change this to the actual dataset folder path
output_folder = "/content/drive/MyDrive/Demand Forecasting (Prince Pharma)/Engineered Data/Forecast Files/Region Level"

# Ensure output folder exists
os.makedirs(output_folder, exist_ok=True)

# Get list of all dataset files
dataset_files = [f for f in os.listdir(dataset_folder) if f.endswith(".csv")]

# Dictionary to store DataFrames grouped by region
region_data = {}

# Loop through each dataset
for dataset in dataset_files:
    dataset_path = os.path.join(dataset_folder, dataset)

    # Read CSV file
    df = pd.read_csv(dataset_path)

    # Check if "region" column exists
    if "region" not in df.columns:
        print(f"⚠️ Skipping {dataset} (No 'region' column)")
        continue

    # Get unique region values (assuming only one per file)
    unique_regions = df["region"].unique()

    for region in unique_regions:
        # Filter data for that specific region
        region_df = df[df["region"] == region]

        # Append to existing data if the region is already seen
        if region in region_data:
            region_data[region] = pd.concat([region_data[region], region_df], ignore_index=True)
        else:
            region_data[region] = region_df

# Save merged datasets for each region
for region, merged_df in region_data.items():
    output_path = os.path.join(output_folder, f"{region}.csv")
    merged_df.to_csv(output_path, index=False)
    print(f"✅ Saved merged dataset for region: {region} -> {output_path}")
```

Region wise Aggregated Execution

```python
import os
import pandas as pd
import re

# Define folder paths
input_folder = "/content/drive/MyDrive/Demand Forecasting (Prince Pharma)/Engineered Data/Forecast Files/Region Level"  # Folder where region-wise datasets are stored
output_folder = "/content/drive/MyDrive/Demand Forecasting (Prince Pharma)/Engineered Data/Forecast Files/Aggregated Region Level"  # Folder to store processed datasets

# Ensure output folder exists
os.makedirs(output_folder, exist_ok=True)

# Function to identify month-year columns
def find_month_year_columns(df):
    month_patterns = [
        r"(?i)^(Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[ -]?(\d{2}|\d{4})$",  # "Jan 24" or "Jan-2024"
        r"(?i)^(January|February|March|April|May|June|July|August|September|October|November|December)[ -]?(\d{2}|\d{4})$",  # "January 24"
        r"(?i)^(JAN|FEB|MAR|APR|MAY|JUN|JUL|AUG|SEP|OCT|NOV|DEC)[ -]?(\d{2}|\d{4})$",  # "JAN 24"
    ]

    month_year_cols = []
    for col in df.columns:
        if any(re.match(pattern, col) for pattern in month_patterns):
            month_year_cols.append(col)

    return month_year_cols

# Process all files in the input folder
for file in os.listdir(input_folder):
    if file.endswith(".csv"):  # Process only CSV files
        file_path = os.path.join(input_folder, file)

        # Load dataset
        df = pd.read_csv(file_path)

        # Extract region name from the dataset
        region_name = df.iloc[0, -1] if isinstance(df.iloc[0, -1], str) else "Unknown_Region"

        # Identify month-year columns
        month_year_cols = find_month_year_columns(df)

        # Group by 'item_id' and sum demand forecasts
        grouped_df = df.groupby("item_id")[month_year_cols].sum().reset_index()

        # Add region name as a new column
        grouped_df["Region"] = region_name

        # Save aggregated dataset
        output_file = os.path.join(output_folder, f"{region_name}_aggregated_demand_forecasts.csv")
        grouped_df.to_csv(output_file, index=False)

        print(f"✅ Processed: {file}, saved as {output_file}")
```

National level product based execution

```python
import os
import pandas as pd
import re
from functools import reduce

# Define input and output folders
input_folder = "/content/drive/MyDrive/Demand Forecasting (Prince Pharma)/Engineered Data/Forecast Files/Aggregated Region Level"  # Update this path
output_folder = "/content/drive/MyDrive/Demand Forecasting (Prince Pharma)/Engineered Data/Forecast Files/National Level"  # Update this path
output_file = os.path.join(output_folder, "nationwide_forecast.csv")

# Ensure output folder exists
os.makedirs(output_folder, exist_ok=True)

# Define month-year patterns
month_patterns = [
    r"(?i)^(Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[ -]?(\d{2}|\d{4})$",  # "Jan 24" or "Jan-2024"
    r"(?i)^(January|February|March|April|May|June|July|August|September|October|November|December)[ -]?(\d{2}|\d{4})$",  # "January 24"
    r"(?i)^(JAN|FEB|MAR|APR|MAY|JUN|JUL|AUG|SEP|OCT|NOV|DEC)[ -]?(\d{2}|\d{4})$",  # "JAN 24"
]

# Load all datasets
dataframes = []
for file in os.listdir(input_folder):
    if file.endswith(".csv"):  # Ensure only CSV files are processed
        file_path = os.path.join(input_folder, file)
        df = pd.read_csv(file_path)
        dataframes.append(df)

# Ensure all datasets have at least one file loaded
if not dataframes:
    raise ValueError("No CSV files found in the input folder!")

# Find common columns across all datasets
common_columns = set(dataframes[0].columns)
for df in dataframes[1:]:
    common_columns.intersection_update(df.columns)

# Keep only relevant columns and merge all datasets
common_columns = list(common_columns)
merged_df = reduce(lambda left, right: pd.merge(left, right, on=common_columns, how="outer"), dataframes)

# Identify columns matching month-year patterns
month_columns = [
    col for col in merged_df.columns
    if any(re.match(pattern, col) for pattern in month_patterns)
]

# Sum the demand product-wise for each month
summary_df = merged_df.groupby("item_id")[month_columns].sum().reset_index()

# Rename columns for better readability
summary_df.columns = ["item_id"] + [f"month_{i+1}" for i in range(len(month_columns))]

# Save the final merged demand summary
summary_df.to_csv(output_file, index=False)

print(f"✅ Processed {len(dataframes)} files, output saved at: {output_file}")
```
